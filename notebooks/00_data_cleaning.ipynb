## Overview
This notebook prepares the raw ATP match data for analysis across all subsequent notebooks in this project.

- **Dataset:** Jeff Sackmann's ATP Match Data

## Why these years?
The 2020 and 2021 seasons were significantly disrupted by the COVID-19 pandemic:
- Wimbledon 2020 was cancelled entirely
- Travel restrictions prevented several players from competing
- Tournaments were played without crowds, affecting player performance
- The competitive environment was different and less representative of normal conditions

Using 2022–2024 ensures we are working with **clean, consistent, and comparable data** from fully operational seasons.

1. Loads and combines raw CSV files into a single dataframe
2. Explores the raw data to understand its structure
3. Handles missing values with documented decisions
4. Fixes data types (date formatting)
5. Engineers new features for analysis and modelling
6. Saves the cleaned dataset to `data/processed/`

```python
# Importing Libraries
import pandas as pd
import numpy as np
import os

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
# Loading raw data
raw_path = '../data/raw/'

df_2022 = pd.read_csv(os.path.join(raw_path, 'atp_matches_2022.csv'))
df_2023 = pd.read_csv(os.path.join(raw_path, 'atp_matches_2023.csv'))
df_2024 = pd.read_csv(os.path.join(raw_path, 'atp_matches_2024.csv'))

# Combine all three years into one dataframe
df = pd.concat([df_2022, df_2023, df_2024], ignore_index=True)

print(f"2022: {len(df_2022)} matches")
print(f"2023: {len(df_2023)} matches")
print(f"2024: {len(df_2024)} matches")
print(f"Total: {len(df)} matches combined")
```

```python
# Check missing values
missing = pd.DataFrame({
    'missing_count': df.isnull().sum(),
    'missing_percent': (df.isnull().sum() / len(df) * 100).round(2)
})

missing = missing[missing['missing_count'] > 0].sort_values('missing_percent', ascending=False)

print(missing)
```

```python
# Dropping columns that are irrelevant to this analysis
cols_not_relevant = [
    'winner_ht', 'loser_ht',
    'winner_hand', 'loser_hand',
    'winner_ioc', 'loser_ioc',
    'tourney_id',
    'match_num',
    'winner_id', 'loser_id',
    'w_ace', 'w_df', 'w_svpt',
    'w_1stIn', 'w_1stWon', 'w_2ndWon',
    'w_SvGms', 'w_bpSaved', 'w_bpFaced',
    'l_ace', 'l_df', 'l_svpt',
    'l_1stIn', 'l_1stWon', 'l_2ndWon',
    'l_SvGms', 'l_bpSaved', 'l_bpFaced'
]

df = df.drop(columns=cols_not_relevant)
print(df.shape)
```

```python
# Dropping columns with too much missing data
cols_high_missing = [
    'winner_entry', 'loser_entry',      # 77-85% missing
    'winner_seed', 'loser_seed',        # 58-75% missing
]

df = df.drop(columns=cols_high_missing)
print(f"Columns after dropping high-missing columns: {df.shape[1]}")
print(df.columns.tolist())
```

```python
# Dropping rows with critical missing values
before = len(df)

# Drop rows where ranking data is missing — we can't analyse without it
df = df.dropna(subset=['winner_rank', 'loser_rank'])

# Drop rows where surface is missing — needed for surface analysis
df = df.dropna(subset=['surface'])

# Drop rows where age is missing — only 2-3 rows, not worth keeping
df = df.dropna(subset=['winner_age', 'loser_age'])

after = len(df)

print(f"Rows before: {before}")
print(f"Rows after: {after}")
print(f"Rows dropped: {before - after}")
```

```python
# Checking 'minutes' column
missing_minutes = df[df['minutes'].isnull()]
print(f"Total missing minutes: {len(missing_minutes)}")
print(f"\nSample of scores where minutes is missing:")
print(missing_minutes['score'].value_counts().head(20))
```

```python
# Check for retirements and walkovers in missing minutes
# Check for walkovers
walkovers = missing_minutes[missing_minutes['score'].str.contains('W/O', na=False)]
print(f"Walkovers: {len(walkovers)}")

# Check for retirements
retirements = missing_minutes[missing_minutes['score'].str.contains('RET', na=False)]
print(f"Retirements: {len(retirements)}")

# Everything else — completed matches with missing duration
completed = missing_minutes[
    ~missing_minutes['score'].str.contains('W/O|RET', na=False)
]
print(f"Completed matches with missing duration: {len(completed)}")
```

```python
# Explore median duration by surface and best_of
median_by_group = df[df['minutes'] > 0].groupby(['surface', 'best_of'])['minutes'].median()
print("Median match duration by surface and best_of:")
print(median_by_group)
```

```python
# Imputing missing values in 'minutes' column
# Fill walkovers with 0 — no time spent on court
df.loc[df['score'].str.contains('W/O', na=False) & df['minutes'].isnull(), 'minutes'] = 0

# Fill retirements with 60% of median by surface and best_of
retirement_mask = df['score'].str.contains('RET', na=False) & df['minutes'].isnull()
df.loc[retirement_mask, 'minutes'] = df.loc[retirement_mask].apply(
    lambda row: round(median_by_group.get((row['surface'], row['best_of']), median_by_group.mean()) * 0.6),
    axis=1
)

# Fill remaining completed matches with median by surface and best_of
def fill_median(row):
    if pd.isnull(row['minutes']):
        return median_by_group.get((row['surface'], row['best_of']), median_by_group.mean())
    return row['minutes']

df['minutes'] = df.apply(fill_median, axis=1)

print(f"Missing minutes remaining: {df['minutes'].isnull().sum()}")
print(f"\nMinutes distribution after imputation:")
print(df['minutes'].describe())
```

## Missing Data Imputation Strategy — Match Duration (`minutes`)

Out of **660 missing values** in the `minutes` column, we identified three distinct categories
by examining the `score` column:

| Category | Count | Imputation Strategy | Reasoning |
|---|---|---|---|
| Walkovers (`W/O`) | 18 | Fill with **0** | Player never took the court — zero time spent playing |
| Retirements (`RET`) | 9 | Fill with **60% of group median** | Match started but did not finish — assumed roughly 60% of a full match was played |
| Completed matches | 633 | Fill with **group median** | Match finished but duration was not recorded |

### Why group median instead of global median?

Rather than using a single global median, we calculated the median match duration
grouped by **surface** and **best_of** (number of sets). This accounts for two
well-known sources of variation in match duration:

- **Surface:** Clay court matches are longer due to slower rallies. Grass is the fastest surface.
- **Best of 5 vs Best of 3:** Grand Slam matches (best of 5) are on average 50–60 minutes
longer than regular tour matches (best of 3).

Using a global median would have misrepresented player workload, directly impacting
the accuracy of our fatigue monitoring analysis.

## Feature Engineering & Data Reshaping

This section transforms the raw match data into a player-centric format suitable
for analysis and machine learning. It involves two major steps:

### Step 1 — Reshape from match format to player format
The raw data has one row per match (winner vs loser). We reshape it to have
**two rows per match** — one for each player — giving each player their own
perspective of the match. This is essential for building an unbiased ML model
with a balanced 50/50 win/loss target variable.

### Step 2 — Feature engineering
We create the following new columns:

| Feature | Description | Used in |
|---|---|---|
| `win` | 1 if player won, 0 if lost | Notebooks 01, 04 |
| `rank_diff` | Player rank minus opponent rank | Notebooks 01, 04 |
| `opponent_rank` | Ranking of the opponent | Notebooks 01, 04 |
| `opponent_rank_points` | Rank points of the opponent | Notebook 04 |
| `match_load_7` | Matches played in last 7 days | Notebook 05 |
| `match_load_14` | Matches played in last 14 days | Notebook 05 |
| `fatigue_score_7` | Normalised 7-day match load (0-1) | Notebook 05 |
| `fatigue_score_14` | Normalised 14-day match load (0-1) | Notebook 05 |

### Why 7 and 14 days for match load?
- **7 days** captures acute fatigue — tiredness from the current tournament week
- **14 days** captures cumulative fatigue — tiredness built up over two weeks,
  relevant for Grand Slams and back-to-back tournaments

### Why a balanced 50/50 dataset is ideal
With exactly 50% wins and 50% losses, the model is forced to learn genuine
patterns rather than exploiting class imbalance. A random baseline would achieve
50% accuracy — our model should comfortably exceed this.

```python
# Reshape from match format to player format
# Shared match columns (same for both players)
shared_cols = [
    'tourney_name', 'tourney_date', 'tourney_level',
    'surface', 'draw_size', 'round', 'best_of', 'minutes', 'score'
]

# Winner perspective
winner_df = df[shared_cols].copy()
winner_df['player_name'] = df['winner_name']
winner_df['player_rank'] = df['winner_rank']
winner_df['player_rank_points'] = df['winner_rank_points']
winner_df['player_age'] = df['winner_age']
winner_df['opponent_rank'] = df['loser_rank']
winner_df['opponent_rank_points'] = df['loser_rank_points']
winner_df['win'] = 1

# Loser perspective
loser_df = df[shared_cols].copy()
loser_df['player_name'] = df['loser_name']
loser_df['player_rank'] = df['loser_rank']
loser_df['player_rank_points'] = df['loser_rank_points']
loser_df['player_age'] = df['loser_age']
loser_df['opponent_rank'] = df['winner_rank']
loser_df['opponent_rank_points'] = df['winner_rank_points']
loser_df['win'] = 0

# Stack both perspectives
df_long = pd.concat([winner_df, loser_df], ignore_index=True)

# Sort by player and date — essential for match_load calculation
df_long = df_long.sort_values(['player_name', 'tourney_date']).reset_index(drop=True)

print(f"Original shape: {df.shape}")
print(f"Reshaped shape: {df_long.shape}")
print(f"\nWin distribution:")
print(df_long['win'].value_counts())
print(f"\nSample rows:")
print(df_long[['player_name', 'tourney_date', 'surface', 'player_rank', 'opponent_rank', 'win']].head(10))
```

```python
df_long.columns.tolist()
```

```python
# Creating rank_diff feature
# rank_diff = player_rank - opponent_rank
# Negative value = player is higher ranked (better) than opponent
# Positive value = player is lower ranked (worse) than opponent
df_long['rank_diff'] = df_long['player_rank'] - df_long['opponent_rank']

print(f"rank_diff sample statistics:")
print(df_long['rank_diff'].describe())
print(f"\nSample:")
print(df_long[['player_name', 'player_rank', 'opponent_rank', 'rank_diff', 'win']].head(10))
```

```python
# Creating match_load_7 and match_load_14
# For each player, count how many matches they played
# in the 7 and 14 days prior to each match (not including the match itself)

def calculate_match_load(group, days):
    load = []
    dates = group['tourney_date'].values
    for i, date in enumerate(dates):
        cutoff = date - pd.Timedelta(days=days)
        count = sum(1 for d in dates[:i] if d >= cutoff)
        load.append(count)
    return load

df_long['match_load_7'] = (
    df_long.groupby('player_name', group_keys=False)
    .apply(lambda g: pd.Series(calculate_match_load(g, 7), index=g.index), include_groups=False)
)

df_long['match_load_14'] = (
    df_long.groupby('player_name', group_keys=False)
    .apply(lambda g: pd.Series(calculate_match_load(g, 14), index=g.index), include_groups=False)
)

print(f"Match load statistics (7 days):")
print(df_long['match_load_7'].describe())
print(f"\nMatch load statistics (14 days):")
print(df_long['match_load_14'].describe())
```

```python
# Creating fatigue_score columns
# Normalise match load to 0-1 scale using min-max normalisation
# 0 = fully fresh, 1 = highest recorded match load in the dataset

df_long['fatigue_score_7'] = (
    df_long['match_load_7'] / df_long['match_load_7'].max()
)

df_long['fatigue_score_14'] = (
    df_long['match_load_14'] / df_long['match_load_14'].max()
)

print(f"Fatigue score (7 day) range: {df_long['fatigue_score_7'].min():.2f} to {df_long['fatigue_score_7'].max():.2f}")
print(f"Fatigue score (14 day) range: {df_long['fatigue_score_14'].min():.2f} to {df_long['fatigue_score_14'].max():.2f}")
print(f"\nSample:")
print(df_long[['player_name', 'tourney_date', 'match_load_7', 'fatigue_score_7', 'match_load_14', 'fatigue_score_14']].head(10))
```

```python
# Final check
print(f"Final shape: {df_long.shape}")
print(f"\nColumns:")
print(df_long.columns.tolist())
print(f"\nMissing values:")
print(df_long.isnull().sum())
print(f"\nSample of final dataframe:")
print(df_long.head())
```

```python
# Save cleaned Data
# Define output path
processed_path = '../data/processed/'

# Save as CSV — used by all analysis notebooks
df_long.to_csv(os.path.join(processed_path, 'matches_cleaned.csv'), index=False)
print("Saved matches_cleaned.csv")

# Save as Excel — demonstrates Excel compatibility
df_long.to_excel(os.path.join(processed_path, 'matches_cleaned.xlsx'), index=False)
print("Saved matches_cleaned.xlsx")

print(f"\nFinal dataset: {df_long.shape[0]} rows x {df_long.shape[1]} columns")
print(f"Saved to: {processed_path}")
```
